# 1. Create a random 4x4 array from an exponential distribution. Refer numpy functions.

```python
import numpy as np

# Parameters for the exponential distribution
scale = 1.0  # Mean of the distribution (Î» = 1/scale)

# Create a 4x4 array from an exponential distribution
exp_array = np.random.exponential(scale=scale, size=(4, 4))

print("4x4 array from exponential distribution:")
print(exp_array)

```

# 2. Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions. Change the number of bins and the plot view to make the visualization better.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
scale = 1.0
size = 100000

# Data generation
exp_data = np.random.exponential(scale=scale, size=(size, 1))

# Plot
plt.hist(exp_data, density=True, bins=250, histtype="step", color="blue", label="Exponential")

plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

# 3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plot
figure = plt.figure(figsize=(8, 6))
ax = figure.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', linewidth=0.1)

# Labels
ax.set_title(r"$Z = X^2 + Y^2$")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()
```

# 4. Calculate the pearson standard coefficient and spearman rank correlation for the [HP, Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the results using heatmaps. Make sure the values are displaced in the heatmap as well.

```python
# Pandas for managing datasets
import pandas as pd
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Keep only the required numeric columns
features = ["HP",	"Attack",	"Defense",	"Sp. Atk",	"Sp. Def",	"Speed"]
df = df[features].dropna()

# Pearson correlation
pearson_corr = df.corr(method='pearson')

# Spearman correlation
spearman_corr = df.corr(method='spearman')

# Plot Pearson heatmap
plt.figure(figsize=(9, 8))
sns.heatmap(pearson_corr, annot=True)
plt.title("Pearson Correlation Heatmap")
plt.show()

# Plot Spearman heatmap
plt.figure(figsize=(9, 8))
sns.heatmap(spearman_corr, annot=True)
plt.title("Spearman Rank Correlation Heatmap")
plt.show()
```
